# Expected Value Simulations

In quant interviews, you will calculate EV for games. This notebook lets you code simulations to *prove* your theoretical math is correct using the Law of Large Numbers.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Set style
plt.style.use('dark_background')
np.random.seed(42)
```

## 1. The "Roll Until You Stop" Game

**Game:** You can roll a 6-sided die up to 3 times. You can stop at any time and take the face value. What is the optimal strategy, and what is your EV?

Let's simulate the optimal strategy we derived: 
- On roll 3, take whatever.
- On roll 2, only stop if $\ge 4$.
- On roll 1, only stop if $\ge 5$.

```python
def simulate_dice_game(num_trials=100000):
    payouts = []
    for _ in range(num_trials):
        # Roll 1
        r1 = random.randint(1, 6)
        if r1 >= 5:
            payouts.append(r1)
            continue
            
        # Roll 2
        r2 = random.randint(1, 6)
        if r2 >= 4:
            payouts.append(r2)
            continue
            
        # Roll 3
        r3 = random.randint(1, 6)
        payouts.append(r3)
        
    return np.mean(payouts)

ev = simulate_dice_game()
print(f"Simulated Expected Value: {ev:.4f}")
print("Theoretical Expected Value: 4.6667")
```

## 2. St. Petersburg Paradox (Truncated)

Assume the casino has a max payout of $1,000,000 (roughly $2^{20}$). Let's simulate the expected value of this game.

```python
def sim_st_petersburg(num_trials=100000, max_bankroll=1000000):
    payouts = []
    for _ in range(num_trials):
        n = 1
        # 0 is Tails, 1 is Heads
        while random.randint(0, 1) == 0:
            n += 1
            
        payout = 2**n
        if payout > max_bankroll:
            payout = max_bankroll
            
        payouts.append(payout)
        
    return np.mean(payouts)

print(f"Simulation (100k trials): The EV is approx: ${sim_st_petersburg():.2f}")
print("Notice how variable the EV is across different simulation batches!")
```
