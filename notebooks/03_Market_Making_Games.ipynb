# Market Making Game Simulator

This simulator attempts to recreate the adverse selection of a live interview. You will quote a market on a hidden value. The "interviewer bot" will hit your bid or lift your offer based on the true value.

```python
import random
import time

def play_market_making():
    print("--- Market Making Game ---")
    print("The secret value is the sum of 100 random numbers drawn from U(0, 10).")
    print("You start with a P&L of $0.\n")
    
    # EV = 100 * 5 = 500.
    true_value = sum([random.uniform(0, 10) for _ in range(100)])
    
    pnl = 0.0
    inventory = 0
    
    for round_num in range(1, 6):
        print(f"\n[Round {round_num}] Current PNL: ${pnl:.2f} | Inventory: {inventory}")
        
        bid_str = input("Enter your BID (or 'q' to quit): ")
        if bid_str.lower() == 'q': break
            
        ask_str = input("Enter your ASK: ")
        
        try:
            bid = float(bid_str)
            ask = float(ask_str)
            if bid >= ask:
                print("Bid must be strictly less than Ask!")
                continue
        except ValueError:
            print("Invalid numbers.")
            continue
            
        # The "Interviewer" Logic (Adverse Selection)
        # The interviewer knows the true value perfectly.
        print("Interviewer is thinking...")
        time.sleep(1)
        
        if true_value > ask:
            print(f">>> Interviewer: 'I BUY from you at {ask}'")
            inventory -= 1  # You are short
            trade_price = ask
            
        elif true_value < bid:
            print(f">>> Interviewer: 'I SELL to you at {bid}'")
            inventory += 1  # You are long
            trade_price = bid
            
        else:
            print(">>> Interviewer: 'I pass. No trade.'")
            trade_price = None
            
        # Mark to market
        if trade_price is not None:
            # If you are short, you want the true value to be lower than ask.
            # If you are long, you want the true value to be higher than bid.
            pass # Pnl is calculated at the end of the game based on final inventory.
            
    print("\n--- Game Over ---")
    print(f"True Value was: {true_value:.2f}")
    
    # Calculate final PnL
    # For every unit long, PNL is TrueValue - PurchasePrice
    # Actually, simpler: we just track cash and inventory.
    # If I sold 1 at 550, Cash = +550. Inv = -1. Final value: 550 - (1 * TrueValue).
    print("Note: A full PNL tracker requires tracking each trade's execution price.")
```
