# Mental Math & Estimation Simulator

This notebook contains a simulated 'Zetamac' game and Fermi problem generators. Run the cells to practice under time pressure.

```python
import random
import time
from IPython.display import clear_output

def play_zetamac_clone(duration_seconds=120):
    operations = ['+', '-', '*', '/']
    score = 0
    start_time = time.time()
    
    print(f"--- Starting Zetamac Clone ({duration_seconds} seconds) ---")
    print("Enter your answer and press Enter. Type 'q' to quit early.")
    time.sleep(2)
    
    while (time.time() - start_time) < duration_seconds:
        op = random.choice(operations)
        if op == '+':
            a = random.randint(2, 100)
            b = random.randint(2, 100)
            ans = a + b
        elif op == '-':
            # make sure a > b so no negative answers usually on zetamac
            a = random.randint(2, 100)
            b = random.randint(2, 100)
            if a < b:
                a, b = b, a
            ans = a - b
        elif op == '*':
            a = random.randint(2, 12)
            b = random.randint(2, 100)
            ans = a * b
        elif op == '/':
            # integer division only for division
            b = random.randint(2, 12)
            ans = random.randint(2, 100)
            a = ans * b
            
        # Prompt
        time_left = int(duration_seconds - (time.time() - start_time))
        if time_left <= 0: break
        
        user_input = input(f"[{time_left}s]  {a} {op} {b} = ")
        
        if user_input.lower() == 'q':
            break
            
        try:
            if int(user_input) == ans:
                score += 1
                print("Correct!\n")
            else:
                print(f"Wrong. Answer was {ans}\n")
        except ValueError:
            print("Please enter a valid number.\n")

    print("\n===========================")
    print(f"Time's up! Final Score: {score}")
    print("===========================")
```
