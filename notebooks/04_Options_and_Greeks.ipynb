# Options & The Greeks

This notebook visualizes the Black-Scholes pricing model and the core Greeks (Delta, Gamma, Theta, Vega).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as si

plt.style.use('dark_background')

def euro_vanilla(S, K, T, r, sigma, option = 'call'):
    # S: spot price
    # K: strike price
    # T: time to maturity
    # r: interest rate
    # sigma: volatility of underlying asset
    
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    
    if option == 'call':
        result = (S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0))
    if option == 'put':
        result = (K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0) - S * si.norm.cdf(-d1, 0.0, 1.0))
        
    return result

def delta(S, K, T, r, sigma, option='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    if option == 'call':
        return si.norm.cdf(d1)
    else:
        return si.norm.cdf(d1) - 1
```

### Visualizing Delta near Expiry

Notice how an option's Delta becomes a step function (0 or 1) as expiration approaches.

```python
S_range = np.linspace(80, 120, 100)
K = 100
r = 0.01
sigma = 0.2

plt.figure(figsize=(10,6))

# 30 days to expiry
deltas_30d = [delta(s, K, 30/365, r, sigma, 'call') for s in S_range]
plt.plot(S_range, deltas_30d, label='T=30 days')

# 1 day to expiry
deltas_1d = [delta(s, K, 1/365, r, sigma, 'call') for s in S_range]
plt.plot(S_range, deltas_1d, label='T=1 day', linestyle='--')

plt.axvline(100, color='red', alpha=0.3, label='Strike')
plt.title('Call Delta Profile as Time to Expiration Decreases')
plt.xlabel('Spot Price')
plt.ylabel('Delta')
plt.legend()
plt.show()
```
