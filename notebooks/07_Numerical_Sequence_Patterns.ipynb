# Numerical Sequence & Pattern Generator

Many trading firms (like Akuna Capital and TradingScreen) test your pattern recognition by firing sequences of numbers at you and asking for the next logical number.

These sequences can be:
- **Arithmetic / Geometric:** Standard constant difference/ratio.
- **Interleaving:** Two entirely separate sequences mashed together `(A1, B1, A2, B2)`.
- **Polynomial:** The difference of the differences is constant.
- **Prime Deltas / Fibonacci:** Non-standard but mathematically rigorous progressions.

Run the generator to practice spotting these under time pressure.

```python
import random
import time

def seq_arithmetic_geometric():
    start = random.randint(2, 10)
    if random.random() > 0.5:
        # Arithmetic
        diff = random.randint(2, 15)
        if random.random() > 0.5: diff = -diff
        seq = [start + i*diff for i in range(6)]
        return seq[:-1], seq[-1], "Arithmetic (constant difference)"
    else:
        # Geometric
        ratio = random.choice([2, 3, 4, 5, -2, -3])
        seq = [start * (ratio**i) for i in range(5)]
        return seq[:-1], seq[-1], "Geometric (constant ratio)"

def seq_polynomial():
    # Difference of differences is constant (e.g., squares + offset)
    a = random.randint(1, 3)
    b = random.randint(-5, 5)
    c = random.randint(-10, 10)
    seq = [(a*(x**2) + b*x + c) for x in range(1, 7)]
    return seq[:-1], seq[-1], "Polynomial (Squares, difference of differnce is constant)"

def seq_interleaving():
    # Two sequences woven together
    startA = random.randint(1, 10)
    diffA = random.randint(2, 10)
    startB = random.randint(50, 100)
    diffB = random.randint(-10, -2)
    
    seqA = [startA + i*diffA for i in range(4)]
    seqB = [startB + i*diffB for i in range(4)]
    
    woven = []
    for i in range(4):
        woven.append(seqA[i])
        woven.append(seqB[i])
        
    return woven[:-1], woven[-1], "Interleaved (Look at every other number!)"
    
def seq_fibonacci_variant():
    a = random.randint(1, 5)
    b = random.randint(1, 5)
    seq = [a, b]
    for _ in range(5):
        seq.append(seq[-1] + seq[-2])
    return seq[:-1], seq[-1], "Fibonacci Variant (Sum of previous two)"

def seq_prime_deltas():
    primes = [2, 3, 5, 7, 11, 13, 17, 19, 23]
    start = random.randint(5, 20)
    seq = [start]
    for i in range(5):
        seq.append(seq[-1] + primes[i])
    return seq[:-1], seq[-1], "Prime Deltas (Difference is the series of prime numbers)"

def play_sequence_game(num_questions=10):
    generators = [
        seq_arithmetic_geometric, 
        seq_polynomial, 
        seq_interleaving, 
        seq_fibonacci_variant,
        seq_prime_deltas
    ]
    
    print("=== NUMERICAL SEQUENCE DRILL ===")
    print("Find the next number in the pattern. Type 'q' to quit.\n")
    
    score = 0
    start_time = time.time()
    
    for i in range(num_questions):
        gen = random.choice(generators)
        seq, ans, explanation = gen()
        
        print(f"Q{i+1}: ", end="")
        for num in seq:
            print(f"{num}, ", end="")
        print("?")
        
        user_input = input("Next number: ")
        if user_input.lower() == 'q':
            break
            
        try:
            if int(user_input) == ans:
                print("✅ Correct!\n")
                score += 1
            else:
                print(f"❌ Wrong! The answer was {ans}.")
                print(f"Pattern Logic: {explanation}\n")
        except ValueError:
            print(f"❌ Invalid input. Answer was {ans}.\n")
            
    elapsed = time.time() - start_time
    print("==============================")
    print(f"Final Score: {score}/{num_questions}")
    print(f"Time taken: {elapsed:.1f} seconds")
    print("==============================")
```

```python
play_sequence_game(10)
```
