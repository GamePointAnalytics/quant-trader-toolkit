# The Optiver "80-in-8" Simulator

Optiver is famous for its 80 in 8 test: 80 math questions in 8 minutes (480 seconds). 
Unlike Zetamac, which only deals with integers, Optiver heavily tests your ability to manipulate **decimals, fractions, and percentages** under extreme time pressure.

### Rules for the real test (and this simulator):
1. **Decimals:** 0.125 * 320, 1.2 / 0.04
2. **Fractions:** 1/3 + 1/6, 3/8 of 64
3. **Percentages:** 15% of 120

Run the cell below to start the simulator. You have exactly 8 minutes. Try to answer as many as possible correctly!

```python
import random
import time
from fractions import Fraction
from IPython.display import clear_output

def generate_decimal_question():
    # Decimals like 0.125, 0.4, 1.2, 0.05 etc.
    decimals = [0.125, 0.25, 0.5, 0.75, 1.5, 2.5, 0.1, 0.2, 0.4, 0.6, 0.8, 1.2, 0.05, 0.15]
    op = random.choice(['*', '/'])
    if op == '*':
        d1 = random.choice(decimals)
        # Choose an integer that plays nicely with d1
        base = random.randint(2, 20)
        if d1 in [0.125, 0.25, 0.5, 0.75]:
            i2 = base * 8
        else:
            i2 = base * 5
        
        if random.random() > 0.5:
            ans = d1 * i2
            return f"{d1} * {i2}", ans
        else:
            d2 = random.choice([0.1, 0.2, 0.4, 0.5])
            ans = d1 * d2
            # Format standard to avoid float precision weirdness (0.1*0.2 = 0.020000000004)
            return f"{d1} * {d2}", round(ans, 4)
            
    elif op == '/':
        # d1 / d2. Make sure it's clean.
        base = random.randint(2, 12)
        d2 = random.choice([0.1, 0.2, 0.04, 0.05, 0.5, 0.25])
        d1 = round(base * d2, 4)
        ans = base
        return f"{d1} / {d2}", float(ans)

def generate_fraction_question():
    fract_bases = [2, 3, 4, 5, 6, 8, 12, 16]
    den = random.choice(fract_bases)
    num = random.randint(1, den-1)
    
    # Form: N/D of X
    mult = random.randint(2, 12)
    X = den * mult
    ans = num * mult
    return f"{num}/{den} of {X}", float(ans)

def generate_percentage_question():
    percents = [5, 10, 15, 20, 25, 30, 40, 50, 75, 150]
    p = random.choice(percents)
    # Clean base
    base = random.randint(1, 15) * 20
    ans = (p / 100) * base
    return f"{p}% of {base}", float(ans)

def generate_addition_subtraction_decimal():
    a = round(random.uniform(0.1, 5.0), 2)
    b = round(random.uniform(0.1, 5.0), 2)
    if random.random() > 0.5:
        return f"{a} + {b}", round(a + b, 2)
    else:
        if a < b:
            a, b = b, a
        return f"{a} - {b}", round(a - b, 2)

def play_optiver_clone():
    print("--- Optiver 80-in-8 Test Simulator ---")
    print("You have 8 minutes to answer up to 80 questions.")
    print("Tips:")
    print("- Enter answers as integers (if whole number) or decimals (e.g. 0.5).")
    print("- For fractions like 1/3, the test usually wants clean integers, so our simulator guarantees clean integer/decimal answers.")
    print("- Type 'q' to quit early.\n")
    
    duration_seconds = 480
    max_qs = 80
    
    score = 0
    attempts = 0
    
    time.sleep(2)
    start_time = time.time()
    
    while attempts < max_qs and (time.time() - start_time) < duration_seconds:
        # Select question type
        q_type = random.choices(
            [generate_decimal_question, generate_fraction_question, generate_percentage_question, generate_addition_subtraction_decimal],
            weights=[40, 20, 20, 20], 
            k=1
        )[0]
        
        q_str, ans = q_type()
        attempts += 1
        
        time_left = int(duration_seconds - (time.time() - start_time))
        if time_left <= 0: 
            break
            
        user_input = input(f"[Q{attempts}/80] [{time_left}s left]  {q_str} = ")
        
        if user_input.lower() == 'q':
            break
            
        try:
            user_float = float(user_input)
            # Allowing small tolerance for floating point mapping
            if abs(user_float - float(ans)) < 1e-4:
                score += 1
                print("Correct!\n")
            else:
                print(f"Wrong. Correct answer was {ans}\n")
        except ValueError:
            print(f"Invalid input. Exact answer was {ans}\n")
            
    print("\n====================================")
    print("TEST FINISHED")
    print(f"Final Score: {score} out of {attempts} attempted.")
    print(f"Total 80-score potential: {score}/{max_qs}")
    if score >= 60:
        print("Performance: OUTSTANDING (Optiver Pass Range)")
    elif score >= 45:
        print("Performance: GOOD (Keep Practicing)")
    else:
        print("Performance: NEEDS WORK (Focus on mental shortcuts)")
    print("====================================")
```
